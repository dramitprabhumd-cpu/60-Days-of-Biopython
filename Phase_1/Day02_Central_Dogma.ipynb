## Project: The Central Dogma (Transcription & Translation)

### ğŸ¯ Objective
Automate the biological process of converting Genetic information into functional Proteins using Biopython's built-in methods.

### ğŸ› ï¸ Key Concepts
* **Transcription:** Converting the "Coding Strand" of DNA into mRNA using `.transcribe()`.
* **Translation:** Converting mRNA (or DNA) into an Amino Acid sequence using `.translate()`.
* **Stop Codons:** Identifying the end of a protein sequence (represented as `*`).

---

### ğŸ““ Notes
* Biopython follows the **Standard Genetic Code** by default.
* If you are working with Mitochondrial DNA, you must specify the table: `dna_seq.translate(table="Vertebrate Mitochondrial")`.

```python
from Bio.Seq import Seq
```

```python
# 1. Starting with coding DNA sequence
dnaSeq = Seq('ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG')
print (f'DNA: {dnaSeq}')

# 2. DNA to mRNA (Transcription)
# In Biopython, this replaces Thymine (T) with Uracil (U)
mRNA = dnaSeq.transcribe()
print(f'mRNA: {mRNA}')

# 3. mRNA to Protein Seq (Tranlation)
# The stop codon (*) is usually represented by an asterisk
proteinSeq = mRNA.translate()
print(f'Protein: {proteinSeq}')

# 4. Biopython allows direct translation from a DNA sequence
proteinSeqDirect = dnaSeq.translate()
print(f'Protein (Direct from DNA): {proteinSeqDirect}')

```
