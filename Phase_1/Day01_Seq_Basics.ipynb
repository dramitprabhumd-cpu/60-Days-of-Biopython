```python
# 2. Import Seq class
from Bio.Seq import Seq
```

```python
# 3. Create a simple DNA sequence
# Note: It's good practice to use uppercase for DNA strings
myDNA = Seq("AGTACACTGGT")
```

```python
# 4. Basic Exploration
print(f'Sequence: {myDNA}')
print(f'Length: {len(myDNA)} bp')
print(f'First Base: {myDNA[0]}')
print(f'Count of A: {myDNA.count('A')}')
```

```python
# 5. Biological Methods
print(f'Complement: {myDNA.complement()}')
print(f'Reverse Complement: {myDNA.reverse_complement()}')
```
