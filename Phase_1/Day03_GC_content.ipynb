## Project: GC Content Analysis

### ðŸŽ¯ Objective
Determine the thermal stability of a DNA sequence by calculating its Guanine-Cytosine (GC) content percentage.

### ðŸ› ï¸ Key Concepts
* **Hydrogen Bonding:** Understanding that G-C pairs have 3 bonds vs. A-T's 2 bonds.
* **Bio.SeqUtils:** Using specialized Biopython modules for common calculations.
* **Accuracy:** Comparing manual count logic with library-verified functions.

---

### ðŸ““ Notes
* **Why it matters:** In PCR (Polymerase Chain Reaction), sequences with higher GC content require higher temperatures to denature.
* **The Function:** We use `gc_fraction()` from `Bio.SeqUtils`. Note that older versions of Biopython used a function called `GC()`, but `gc_fraction` is the modern standard.

```python
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction

# 1. Define a sequence from a heat-resistant bacteria
stressDna = Seq('GCCGGGCACTCGGCTACTAAAAGGGGCGCGCGAG')

# 2. Manual Calculation
cCount = stressDna.count('C') # Nos. of C
gCount = stressDna.count('G') # Nos. of G
manualGC = (cCount + gCount)/len(stressDna) * 100
print(f'Manual GC percentage = {manualGC:.2f}%')

# 3. Biopython Way
# gc_fration returns a float between 0 & 1
biopythonGC = gc_fraction(stressDna) * 100

print(f'Biopython GC Content = {biopythonGC:.2f}%')

# 4. Quick Logical Check
if biopythonGC >= 50:
  print('This is GC-rich sequence (Likely Stable).')
else:
  print('This is AT-rich sequence')
```
